# Heatmap
Code used to generate the heatmap

```python
# imports
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

Load the dataset

```python
loaded_df = pd.read_csv("../Datasets/Master_Dataset_Detoxified_Complete.csv")
```

Keep only the columns needed for heat map

```python
heat_map = loaded_df[['Hate Speech Flag', 'is_sarcastic', 'toxicity', 'severe_toxicity', 'obscene', 'identity_attack', 'insult',
       'threat', 'sexual_explicit']]

heat_map = heat_map.rename(columns={'is_sarcastic': 'Sarcasm Flag'})
```

Generate heat map using Seaborn

```python
plt.figure(figsize=(12,10))

plt.xticks(rotation=45, ha='right')
plt.subplots_adjust(bottom=0.25)

cor = heat_map.corr()
sns.heatmap(cor, annot=True, cmap=plt.cm.Reds)
plt.show()
```
