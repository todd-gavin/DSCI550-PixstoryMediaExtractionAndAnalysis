```python
import pandas as pd
```

```python
# Prepping object data for merge

# Load JSON file into a Pandas DataFrame
df_objects = pd.read_json('../4_Tika Image Dockers/image_object_recognition_master.json')
print(len(df_objects))

# Define a lambda function to modify the URLs
normalize_url = lambda url: 'https://image.pixstory.com/optimized/' + url.split('/')[-1]

# Apply the lambda function to the URL column
df_objects['url'] = df_objects['url'].apply(normalize_url)

# remove "/optimized" substring from URL
df_objects['url'] = df_objects['url'].str.replace('/optimized', '')

df_objects


#urls should look like this: https://image.pixstory.com/Pixstory-image-165942680078373.jpeg
```

```python
# prepping captions data for merge

df_captions = pd.read_json('../4_Tika Image Dockers/image_captions_data_master.json')

# Define a lambda function to modify the URLs
normalize_url = lambda url: 'https://image.pixstory.com/optimized/' + url.split('/')[-1]

# Apply the lambda function to the URL column
df_captions['url'] = df_captions['url'].apply(normalize_url)

# remove "/optimized" substring from URL
df_captions['url'] = df_objects['url'].str.replace('/optimized', '')

#urls should look like this: https://image.pixstory.com/Pixstory-image-165942680078373.jpeg


df_captions

# print(len(df_captions))
```

```python
#prepparing Pixstory df to merge

ps_df = pd.read_csv('Master_Dataset_Raw.csv')

# Rename column to prepare for merge
ps_df = ps_df.rename(columns={'Media': 'url'})

pd.set_option('display.max_colwidth', 150)

##How to add "/optimized":
#modify_url = lambda url: url.replace(".com/", ".com/optimized/")

## Apply the lambda function to the "Media" column of the DataFrame
#ps_df['Media'] = ps_df['Media'].apply(modify_url)


print(len(ps_df))
ps_df

```

```python
#Merge 1: objects with pixstory

objects_ps_df = pd.merge(df_objects, ps_df, on='url', how='outer')

print(len(ps_df))
print(len(objects_ps_df))

#Merge 2: captions with objects & pixstory
objects_captions_ps_df = pd.merge(objects_ps_df, df_captions, on='url', how='outer')

print(len(objects_captions_ps_df))

objects_captions_ps_df
```
