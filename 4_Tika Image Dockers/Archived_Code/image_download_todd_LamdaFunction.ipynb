```python
import pandas as pd
import urllib.request
import os
import urllib.error
```

```python
# Read the CSV file and create a DataFrame
df = pd.read_csv('../Datasets/Master_Dataset_Raw.csv')
```

```python
# This line of code sets the maximum width of a column in a pandas DataFrame to be displayed without any truncation. 
pd.set_option('display.max_colwidth', None)

# Drop the duplicate URLs from the DataFrame
df = df.drop_duplicates(subset='Media')

# reset the index
df = df.reset_index()

# Define a lambda function to modify the URLs
modify_url = lambda url: url.replace(".com/", ".com/optimized/")

# Apply the lambda function to the "Media" column of the DataFrame
df['Media'] = df['Media'].apply(modify_url)
```

```python
start_index = 6000
end_index = 0
```

```python
def download_image(url):

    global start_index
    global end_index

    current_index = df[df['Media'] == url].index[0]

    print("****************************")
    print("Start Index:", start_index, "\tEnd Index:", end_index, "\tCurrent Index:", current_index)

    if current_index < start_index:
        print(f"Skipping image: {url}")

    else:
        try:
            urllib.request.urlretrieve(url, f"Pixstory_Image_Files/{url.split('/')[-1]}")
            print(f"Downloading image: {url}")

        except Exception as e:
            print(f"Error downloading image: {url} - {str(e)}")

    end_index = current_index
```

```python
# Apply the lambda function to each element in the "Media" column
df['Media'].apply(download_image)
```
